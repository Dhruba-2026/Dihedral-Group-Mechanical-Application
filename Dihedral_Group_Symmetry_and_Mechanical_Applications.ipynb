# 2. Literature Review

Symmetry plays a fundamental role in mathematics, physics, and engineering by providing a systematic framework for understanding the structural and geometric properties of objects. Among the various symmetry groups, **Dihedral Groups** have attracted significant attention due to their ability to describe the rotational and reflectional symmetries of regular polygons and numerous engineering structures. The mathematical formulation of dihedral groups has been extensively studied in the field of abstract algebra and has found applications in crystallography, molecular chemistry, robotics, computer graphics, and mechanical engineering **[1, 2, 3]**.

The concept of group theory was formally established through the pioneering works of mathematicians such as Évariste Galois and later expanded by researchers including Joseph Gallian and John B. Fraleigh. These studies demonstrated that algebraic groups provide a rigorous framework for representing symmetry transformations and analyzing invariant properties of geometric objects **[1, 2]**. In particular, the dihedral group (D_n), consisting of (n) rotational and (n) reflection operations, serves as one of the most important examples of finite groups with practical engineering relevance **[2, 3]**.

In mechanical engineering, symmetry has been widely exploited to simplify computational models and reduce analysis time. Structures such as turbine rotors, gears, impellers, polygonal couplings, and cyclically symmetric assemblies often exhibit dihedral symmetry. By recognizing these symmetry properties, engineers can analyze only a representative sector of the structure and reconstruct the complete solution through appropriate symmetry transformations. This approach significantly reduces computational cost while maintaining solution accuracy and has become a standard practice in finite element analysis (FEA) and computational mechanics **[4, 5]**.

Research in computational structural analysis has further shown that symmetry can be used to reduce the number of independent variables in stress and strain calculations. Tensor transformation laws demonstrate that stress tensors remain invariant or transform predictably under orthogonal symmetry operations. Consequently, exploiting symmetry enables more efficient simulations of elastic deformation, vibration analysis, and structural optimization **[4, 5, 6]**.

Recent developments in computational mathematics have integrated group theory with programming languages such as Python to create educational and research-oriented simulation tools. Libraries including **NumPy** and **Matplotlib** provide efficient numerical computation and visualization capabilities, allowing researchers to implement rotation matrices, reflection matrices, and geometric transformations with relatively simple code. Such implementations facilitate interactive demonstrations of abstract mathematical concepts while simultaneously connecting them with practical engineering applications.

Despite the widespread use of symmetry in commercial finite element software, relatively few educational projects explicitly combine the mathematical foundations of dihedral groups with computational demonstrations of mechanical properties such as moment of inertia, stress tensor transformation, and symmetry-based domain reduction. Bridging this gap offers valuable insights into both theoretical group theory and its engineering applications **[4, 5, 7]**.

The present project addresses this objective by developing a Python-based computational framework that generates dihedral group symmetry operations, visualizes rotational and reflectional transformations of mechanical components, evaluates moment of inertia under symmetric configurations, demonstrates stress tensor transformations, and illustrates the principle of symmetry-based computational domain reduction. The project thereby establishes a clear connection between abstract algebra and practical mechanical engineering analysis through numerical simulation and visualization.



## 3. Theoretical Background

### 3.1 Dihedral Group Dₙ
The Dihedral group Dₙ consists of:
- n rotations: R_k = rotation by (2πk/n)
- n reflections: S_k = reflection about symmetry axes

Total elements:
|Dₙ| = 2n

### 3.2 Rotation Matrix
Rotations in 2D are represented as:

R(θ) = [[cosθ, -sinθ],
        [sinθ, cosθ]]

### 3.3 Reflection Matrix
Reflections are defined as:

S(θ) = [[cos2θ, sin2θ],
        [sin2θ, -cos2θ]]

### 3.4 Mechanical Relevance
These transformations preserve:
- distances
- angles
- physical invariants such as inertia and stress eigenstructure

# 6. Methodology


The methodology adopted in this project combines concepts from **abstract algebra**, **computational mathematics**, and **mechanical engineering** to investigate the role of Dihedral Group symmetry in engineering applications. The complete workflow consists of mathematical modelling, numerical implementation, visualization, and engineering analysis using Python.

The project begins with the mathematical definition of the Dihedral Group \(D_n\), which represents the symmetry group of a regular polygon with \(n\) sides. The group consists of \(n\) rotational symmetries and \(n\) reflection symmetries, giving a total of \(2n\) unique symmetry operations.

For computational implementation, the rotational and reflection operations are represented as transformation matrices. Rotation matrices are generated for angles

$$
[
\theta=\frac{2\pi k}{n},
\qquad k=0,1,\ldots,n-1
]
$$
while reflection matrices are constructed with respect to appropriate symmetry axes. These matrices serve as the fundamental operators for transforming two-dimensional mechanical geometries.

A representative mechanical component, modelled as a simplified rotor blade geometry, is then defined in Cartesian coordinates. Each symmetry operation belonging to the Dihedral Group is applied to the original geometry, producing multiple transformed copies and thereby reconstructing the complete symmetric structure. This process provides a visual demonstration of rotational and reflectional invariance.

The transformed geometries are subsequently utilized for engineering analysis. A discrete formulation of the **moment of inertia** is implemented to evaluate the rotational characteristics of the complete assembly. The project further demonstrates that, owing to cyclic symmetry, the inertia of the full structure can be estimated efficiently by computing the inertia of a single sector and scaling the result appropriately.

To investigate tensorial quantities under symmetry operations, a two-dimensional **stress tensor transformation** is performed. Using orthogonal transformation matrices, the stress components are mapped into new coordinate systems according to tensor transformation laws. This simulation illustrates the preservation of physical consistency under symmetry transformations and highlights the role of group operations in continuum mechanics.

An additional computational experiment demonstrates the concept of **symmetry-based domain reduction**, a widely used technique in finite element analysis (FEA). Instead of analysing the entire geometry, only a representative sector is considered while exploiting the inherent symmetry of the structure. This approach reduces computational cost and memory requirements without sacrificing solution accuracy.

Finally, graphical visualization techniques implemented using **Matplotlib** are employed to display the transformed geometries, symmetry operations, and computational results. Animated visualizations further illustrate rotational motion and provide an intuitive understanding of Dihedral Group actions on mechanical systems.

The overall methodology establishes a systematic connection between theoretical group theory and practical engineering analysis through mathematical modelling, numerical simulation, and scientific visualization.

## 7. Python code Implementation
The following sections implement the mathematical concepts developed using Python.


`````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import distance_matrix

# Set up interactive plots for Jupyter if desired
%matplotlib inline

# =====================================================================
# 7.1. MATHEMATICAL MODELING OF DIHEDRAL GROUP (D_n)
# =====================================================================

class DihedralGroup:
    """
    Models the Dihedral Group D_n representing the symmetries of a regular n-gon.
    Contains 2n elements: n rotations and n reflections.
    """
    def __init__(self, n):
        self.n = n
        self.alpha = 2 * np.pi / n  # Fundamental rotation angle
        self.elements = self._generate_group_matrices()

    def _generate_group_matrices(self):
        """Generates the 2D transformation matrices for all 2n elements."""
        matrices = {}

        # Rotations (R_k)
        for k in range(self.n):
            theta = k * self.alpha
            R = np.array([
                [np.cos(theta), -np.sin(theta)],
                [np.sin(theta),  np.cos(theta)]
            ])
            matrices[f'R_{k}'] = R

        # Reflections (S_k) across axes tilted by k * alpha / 2
        for k in range(self.n):
            phi = k * (self.alpha / 2)
            S = np.array([
                [np.cos(2*phi),  np.sin(2*phi)],
                [np.sin(2*phi), -np.cos(2*phi)]
            ])
            matrices[f'S_{k}'] = S

        return matrices

    def get_base_polygon(self):
        """Generates vertices of a regular n-gon inscribed in a unit circle."""
        angles = [i * self.alpha for i in range(self.n)]
        return np.array([[np.cos(a), np.sin(a)] for a in angles])

    def plot_symmetry_operations(self):
        """Visualizes how rotation and reflection alter a test point."""
        polygon = self.get_base_polygon()
        test_point = np.array([0.8, 0.2])  # Asymmetric point to show transformation clearly

        fig, axes = plt.subplots(1, 2, figsize=(12, 5))

        # Plot 1: Rotations
        ax = axes[0]
        ax.plot(np.append(polygon[:,0], polygon[0,0]), np.append(polygon[:,1], polygon[0,1]), 'k--', alpha=0.5)
        ax.scatter(polygon[:,0], polygon[:,1], color='gray', s=100)

        for name, matrix in self.elements.items():
            if name.startswith('R'):
                pt = matrix @ test_point
                ax.scatter(pt[0], pt[1], label=name, s=80)
                ax.text(pt[0]*1.1, pt[1]*1.1, name, fontsize=9)
        ax.set_title(f"Rotation Operations ($R_k$) in $D_{self.n}$")
        ax.axis('equal')
        ax.grid(True)

        # Plot 2: Reflections
        ax = axes[1]
        ax.plot(np.append(polygon[:,0], polygon[0,0]), np.append(polygon[:,1], polygon[0,1]), 'k--', alpha=0.5)
        ax.scatter(polygon[:,0], polygon[:,1], color='gray', s=100)

        for name, matrix in self.elements.items():
            if name.startswith('S'):
                pt = matrix @ test_point
                ax.scatter(pt[0], pt[1], label=name, s=80, marker='X')
                ax.text(pt[0]*1.1, pt[1]*1.1, name, fontsize=9)
        ax.set_title(f"Reflection Operations ($S_k$) in $D_{self.n}$")
        ax.axis('equal')
        ax.grid(True)

        plt.tight_layout()
        plt.show()

# =====================================================================
# 7.2. MECHANICAL APPLICATIONS
# =====================================================================

class MechanicalSymmetryAnalyzer:
    def __init__(self, group: DihedralGroup):
        self.group = group

    def calculate_rotor_moment_of_inertia(self, base_blade_particles, particle_mass):
        """
        Models a symmetric mechanical rotor by populating an initial sector (blade)
        across all symmetric orientations of the D_n group.
        Computes the total Mass Moment of Inertia (I_z).
        """
        total_I_z = 0.0
        all_particles = []

        # Apply only unique rotational parts to avoid double-counting if reflections overlap
        # For a standard rotor, we replicate the base blade via the N rotational sectors
        for name, matrix in self.group.elements.items():
            if name.startswith('R'):  # Replicate blade across sectors
                for particle in base_blade_particles:
                    transformed_pt = matrix @ particle
                    all_particles.append(transformed_pt)

                    # I_z = sum(m * r^2) where r^2 = x^2 + y^2
                    r_squared = np.sum(transformed_pt**2)
                    total_I_z += particle_mass * r_squared

        return total_I_z, np.array(all_particles)

    def transform_stress_tensor(self, stress_tensor_2d, operation_name):
        """
        Transforms a 2D Cauchy Stress Tensor using the group representation matrix.
        Formula: sigma' = Q * sigma * Q^T
        """
        Q = self.group.elements[operation_name]
        transformed_stress = Q @ stress_tensor_2d @ Q.T
        return transformed_stress

    def analyze_domain_reduction(self):
        """
        Demonstrates the domain reduction factor. In FEA, instead of analyzing
        the whole object, we only analyze 1/(2n) of the domain if reflections are used.
        """
        domain_fraction = 1 / (2 * self.group.n)
        reduction_percentage = (1 - domain_fraction) * 100
        print(f"--- Computational Domain Reduction Analysis ($D_{self.group.n}$) ---")
        print(f"Symmetry Group Size (Order): {2 * self.group.n}")
        print(f"Minimum required FEA analysis domain: {domain_fraction*100:.2f}% of total structure.")
        print(f"Computational resource reduction factor: {reduction_percentage:.2f}% fewer elements needed.")
        print("-" * 65)


# =====================================================================
# 7.3. CRYSTALLOGRAPHIC MAPPING (Link to Materials Science)
# =====================================================================

def analyze_crystal_lattice_symmetry(n_fold=4):
    """
    Simulates a square (4-fold) or hexagonal (6-fold) 2D crystal lattice plane
    and verifies if the stress tensor remains invariant under specific D_n mappings.
    """
    print(f"\n[Materials Science] Testing Stress Invariance in a {n_fold}-fold Crystallographic Sheet")
    dg = DihedralGroup(n_fold)
    analyzer = MechanicalSymmetryAnalyzer(dg)

    # Define an isotropic/hydrostatic stress state vs shear stress state
    hydrostatic_stress = np.array([[200, 0],
                                    [0, 200]]) # MPa

    pure_shear_stress = np.array([[0, 150],
                                  [150, 0]]) # MPa

    # Test invariance under a fundamental rotation R_1
    trans_hydro = analyzer.transform_stress_tensor(hydrostatic_stress, 'R_1')
    trans_shear = analyzer.transform_stress_tensor(pure_shear_stress, 'R_1')

    print("Original Hydrostatic Stress:\n", hydrostatic_stress)
    print(f"Transformed Hydrostatic Stress under R_1:\n", np.round(trans_hydro, 2))
    print(">> Hydrostatic stress is invariant under crystal rotation operations!")

    print("\nOriginal Pure Shear Stress:\n", pure_shear_stress)
    print(f"Transformed Pure Shear Stress under R_1:\n", np.round(trans_shear, 2))


# =====================================================================
# 7.4. EXECUTION PIPELINE (Jupyter Execution)
# =====================================================================

if __name__ == "__main__":
    # Let's model a 6-bladed mechanical rotor / Hexagonal crystal symmetry (D_6)
    N_SYMMETRY = 6
    print(f"Initializing Simulation for Dihedral Group D_{N_SYMMETRY}...\n")

    # Step 1: Initialize Math Group
    d6_group = DihedralGroup(n=N_SYMMETRY)

    # Step 2: Show Mathematical Symmetry Operations
    d6_group.plot_symmetry_operations()

    # Step 3: Connect Math with Engineering - Mechanical Rotor Analysis
    analyzer = MechanicalSymmetryAnalyzer(d6_group)

    # Define a single rotor blade point-cloud (x, y) in the first sector
    # A line of mass particles extending outward along the x-axis
    base_blade_mass_points = np.array([
        [0.2, 0.0], [0.4, 0.0], [0.6, 0.0], [0.8, 0.0], [1.0, 0.0],
        [0.4, 0.05], [0.6, 0.05], [0.8, 0.05] # thickness components
    ])
    particle_mass_kg = 0.1 # 100 grams per node

    total_I_z, full_rotor_mesh = analyzer.calculate_rotor_moment_of_inertia(
        base_blade_mass_points, particle_mass_kg
    )

    print(f"Total Polar Moment of Inertia (I_z) for the generated {N_SYMMETRY}-blade Rotor: {total_I_z:.5f} kg·m²\n")

    # Plotting the generated symmetric mechanical system
    plt.figure(figsize=(6,6))
    plt.scatter(full_rotor_mesh[:,0], full_rotor_mesh[:,1], c='crimson', s=50, edgecolors='k', label='Symmetric Mass Nodes')
    # Plot center of rotation
    plt.scatter(0, 0, c='black', marker='+', s=200, label='Hub Center')
    plt.title(f"Symmetrically Generated {N_SYMMETRY}-Blade Rotor Mesh via $D_6$")
    plt.xlabel("X-coordinate (m)")
    plt.ylabel("Y-coordinate (m)")
    plt.axis('equal')
    plt.grid(True)
    plt.legend()
    plt.show()

    # Step 4: Domain Reduction Factor Analysis
    analyzer.analyze_domain_reduction()

    # Step 5: Stress Tensor Transformations & Crystallographic Mapping
    analyze_crystal_lattice_symmetry(n_fold=4)
`````
